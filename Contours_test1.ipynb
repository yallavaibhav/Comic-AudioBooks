```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#uploading the image
src = cv2.imread('/content/Screenshot 2023-02-12 233231.png')
plt.imshow(src)
```

```python
#Gray Scalling
src_gray = cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
src_gray = cv2.blur(src_gray, (3,3))
plt.imshow(src_gray)
```

```python
#Threshoding
ret, threshold = cv2.threshold(src_gray,10,255,cv2.THRESH_BINARY)
```

```python
contours, hir = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
hull = [cv2.convexHull(c) for c in contours]
final = cv2.drawContours(src,hull,-1, (255,225,255))
plt.imshow(final)
```

Canny Edge Detection

```python
#Rerun the entire program (restart), then do't run the threshold
edges = cv2.Canny(src_gray,100,200)
```

```python
contours, hir = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
hull = [cv2.convexHull(c) for c in contours]
final = cv2.drawContours(src,hull,1, (255,225,255))
plt.imshow(final)
```
